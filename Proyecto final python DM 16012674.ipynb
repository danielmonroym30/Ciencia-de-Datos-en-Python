```python
def regresion(epoch,lr,x):   #EXITO
    error_hist=np.empty((epoch,2))
    modelo_parametros=np.empty((epoch,2))
    m=0
    b=0
    
    for i in range(0,epoch):
        parametros=np.array([[m],[b]])
        yhat=np.matmul((np.array([datos_entrenamiento[:,x],unos])).T,parametros)  #
        y=datos_entrenamiento[:,0]
        error=1/2*(np.mean((y-yhat)**2))
        error_hist[i]=[i,error]
        gradm=np.mean((yhat-y)*(np.array([datos_entrenamiento[:,x],unos])).T[:,0])
        gradb=np.mean((yhat-y)*1)
        gradientes=np.hstack((gradm,gradb))
        m=m-(lr*gradm)
        b=b-(lr*gradb)
        m_b=np.hstack((m,b))
        modelo_parametros[i]=[m,b]
        frame=np.append(error_hist,modelo_parametros,axis=1)
        dataframe=pd.DataFrame(frame,columns=['Iteración','Error','m','b'])
        
    return dataframe


regresion_lineal1=regresion(100,0.001,1) #"1" debido a que es la primera variable x a entrenar
columna_iteracion1=regresion_lineal1["Iteración"]
columna_error1=regresion_lineal1["Error"]
columna_mb1=regresion_lineal1[["m","b"]]


regresion_lineal1

                                                   
```
